```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
miles = pd.read_csv('us-airlines-monthly-aircraft-miles-flown.csv', header = 0, parse_dates=[0])
```

```python
miles.head()
```

```python
miles['MilesMM'].plot()
```

```python
miles['lag1'] = miles['MilesMM'].shift(1)
```

```python
miles['diff1'] = miles['MilesMM']-miles['lag1']
```

```python
miles['diff_1'] = miles['MilesMM'].diff(periods=1)
```

```python
miles
```

```python
miles['diff_1'].plot()
```

```python
miles.index = miles['Month']

seasonal_decompose(miles['diff1'].iloc[1:], model='Additive').plot()
```

```python
miles['diff12'] = miles['diff1'].diff(12)
```

```python
seasonal_decompose(miles['diff12'].iloc[13:], model='Additive').plot()
```
