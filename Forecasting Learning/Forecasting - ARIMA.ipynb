# ARIMA

* ## AR is used for Forecasting the values, MA is used for Forecasting the residuals and Integrated refers to differencing.

* ## Differencing handles Trend but not seasonality.



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf, plot_predict
from pandas.plotting import autocorrelation_plot
```

```python
df = pd.read_csv('shampoo.csv', header = 0, parse_dates=[0])
```

```python
df['Sales'].plot()
```

* Trend is quadratic, hence d = 2

```python
autocorrelation_plot(df['Sales'])
```

```python
plot_acf(df['Sales'], lags = 10)
```

* MA - q = 3

```python
plot_pacf(df['Sales'], lags = 5)
```

* AR - p = 2

```python
model = ARIMA(df["Sales"], order = (2, 2, 3))
```

```python
model_fit = model.fit()
```

```python
model_fit.summary()
```

```python
float(model_fit.forecast())
```

# Walk forward Validation

```python
df = pd.read_csv('shampoo.csv', header = 0 , parse_dates=[0])
```

```python
df.head()
```

```python
train_size = int(df.shape[0]*0.7)
```

```python
train, test = df.iloc[:train_size], df.iloc[train_size:]
```

```python
test.shape, train.shape
```

```python
data = train['Sales']
predictions = []

for t in test['Sales']:
    model = ARIMA(data, order = (2,2,3))
    model_fit = model.fit()
    predictions.append(float(model_fit.forecast()))
    data = np.append(data, t)
    data = pd.Series(data)
```

```python
plt.plot(predictions)
plt.plot(test['Sales'].values)
```
