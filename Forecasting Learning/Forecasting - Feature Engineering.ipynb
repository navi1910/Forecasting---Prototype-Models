```python
import pandas as pd

import numpy as np

data = pd.read_csv('daily-total-female-births-CA.csv', header = 0, parse_dates=[0])
```

```python
df = data.copy()
```

```python
df
```

# Date Time Features

```python
df['Year'] = df['date'].dt.year
```

```python
df['Month'] = df['date'].dt.month
```

```python
df['Day'] = df['date'].dt.day
```

```python
df
```

# Lag Features

```python
#Info of the Previous Time Period

df['Lag'] = df['births'].shift(1)

# Info of same day last year

df['Lag2'] = df['births'].shift(365)
```

```python
df
```

```python
df['Rolling_mean2'] = df['births'].rolling(window=2).mean()
```

```python
df['Rolling_max3'] = df['births'].rolling(window=3).max()
```

```python
df
```

# Expanding Windows

```python
df['Expand_Max'] = df['births'].expanding().max()
```
