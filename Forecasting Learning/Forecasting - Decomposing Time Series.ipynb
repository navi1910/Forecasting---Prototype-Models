```python
import pandas as pd
import numpy as np

from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
miles = pd.read_csv('us-airlines-monthly-aircraft-miles-flown.csv', header = 0, parse_dates= [0])
```

```python
miles.head()
```

```python
miles.index = miles['Month']
```

```python
add = seasonal_decompose(miles['MilesMM'], model='Additive')
```

```python
add.plot()
```

```python
multi = seasonal_decompose(miles['MilesMM'], model='Multiplicative')
```

```python
multi.plot()
```
