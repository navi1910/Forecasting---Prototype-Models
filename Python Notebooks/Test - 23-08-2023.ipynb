```python
orders(df['Sales_Value'], lags = 10, duration=30)
```

```python
df['Lag7'] = differencing(df['Sales_Value'], lag=7)
```

```python
arima_1 = arima(df['Lag7'].iloc[7:], p = 2, d = 0, q = 3, summary=True)
```

```python
tr, ts = train_test_split(df['Lag7'].iloc[7:], train_size=0.8)
```

```python
pred = arima_1.predict(start = '2022-12-24', end = '2023-06-28')
```

```python
plt.plot(ts.loc[:'2023-06-28'], c = 'g', alpha = 0.7)

plt.plot(pred.loc[:'2023-06-28'], c = 'r',  ls = '--', alpha = 0.6)
```

```python
data = tr
predictions = []

for t in ts:
    model = ARIMA(data, order = (2,0,3))
    model_fit = model.fit()
    predictions.append(float(model_fit.forecast()))
    data = np.append(data, t)
    data = pd.Series(data)
```
