```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.seasonal import seasonal_decompose
import pandas as pd
```

```python
df = pd.read_csv('us-airlines-monthly-aircraft-miles-flown.csv', header = 0, parse_dates = [0])
```

```python
df.head()
```

```python
df.index = df['Month']
```

```python
seasonal_decompose(df['MilesMM'], model = 'multiplicative').plot()
```

```python
model = SARIMAX(df['MilesMM'], order = (5,1,3), seasonal_order=(1,1,1,12)).fit()
```

```python
model.resid.plot()
```

```python
output = model.forecast()
```

```python
output
```

```python
yhat = model.predict()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(yhat)
plt.plot(df['MilesMM'])
```
