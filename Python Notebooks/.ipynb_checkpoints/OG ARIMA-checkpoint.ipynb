```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('Sales Forecasting data.csv', index_col=[0])
```

```python
df.head()
```

```python
df['Sales_Value'].plot()
```

```python
df['Lag7'] = df['Sales_Value'].shift(7)
```

```python
df['Lag7'].head(8)
```

```python
data = df['Lag7'].iloc[7:]
```

```python
plot_acf(data, lags = 10)
```

```python
plot_pacf(data, lags = 10)
```

```python
train_size = int(data.shape[0]*0.80)
```

```python
train, test = data.iloc[:train_size], data.iloc[train_size:]
```

```python
test.shape[0]
```

```python
model = ARIMA(data, order = (1, 0, 2)).fit()
```

```python
model.summary()
```

```python
pred = model.predict(start = len(train), end = len(train)+len(test)-1)
```

```python
pred
```

```python
plt.plot(pred.values, c='r', ls = '--', label = "Preds",  alpha = 0.8)
plt.plot(test.values, alpha = 0.3, label = 'Actual')
plt.legend()
```

```python
np.sqrt(mean_squared_error(pred, test.values))
```

# Walk Forward Validation

```python
tdata = train
walk_pred = []

for t in test:
    model = ARIMA(tdata, order = (1,0,2))
    model_fit = model.fit()
    walk_pred.append(float(model_fit.forecast()))
    tdata = np.append(tdata, t)
    tdata = pd.Series(tdata)
```

```python
plt.plot(walk_pred, c='r', ls = '--', label = "Preds",  alpha = 0.8)
plt.plot(test.values, alpha = 0.3, label = 'Actual')
plt.legend()
```

```python
np.sqrt(mean_squared_error(walk_pred, test.values))
```

# SARIMAX

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
tdata = train
walk_pred = []

for t in test:
    model = SARIMAX(tdata, order = (1,0,2), seasonal_order=(1,1,1,7))
    model_fit = model.fit()
    walk_pred.append(float(model_fit.forecast()))
    tdata = np.append(tdata, t)
    tdata = pd.Series(tdata)
```

```python
plt.plot(walk_pred, c='r', ls = '--', label = "Preds",  alpha = 0.8)
plt.plot(test.values, alpha = 0.3, label = 'Actual')
plt.legend()
```

```python
np.sqrt(mean_squared_error(walk_pred, test.values))
```
