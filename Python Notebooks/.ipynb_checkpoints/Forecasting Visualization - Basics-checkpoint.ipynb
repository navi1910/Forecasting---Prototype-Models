```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
 data = pd.read_csv('daily-total-female-births-CA.csv', header = 0, parse_dates=[0])
```

```python
df = data.copy()

df.head()
```

```python
df['births'].plot()
```

```python
df.index = df['date']
```

```python
df['births'].plot()
```

# Zooming In

```python
df_jan = df[(df['date'] >= '1959-01-01') & (df['date'] <= '1959-01-31')]
```

```python
df_jan['births'].plot()
```

# Trend Line

```python
df = data.copy()

sns.regplot(x = df.index.values, y = df['births']) # Linear Trend Line
```

```python
sns.regplot(x = df.index.values, y = df['births'], order = 2) # Quadratic Trend Line
```

```python
miles = pd.read_csv('us-airlines-monthly-aircraft-miles-flown.csv', header = 0, parse_dates=[0])
```

```python
miles.head()
```

```python
miles['MilesMM'].plot()
```

```python
sns.regplot(x = miles.index.values, y = miles['MilesMM'], order = 1)
```

# Removing Seasonality

```python
miles['Year'] = miles['Month'].dt.year
```

```python
miles
```

```python
miles.groupby('Year')['MilesMM'].mean().plot()
```

# Lag Plots

```python
miles['Lag1'] = miles['MilesMM'].shift(1)
```

```python
miles.head()
```

```python
sns.regplot(x = miles['MilesMM'], y = miles['Lag1'])
```

```python
from pandas.plotting import lag_plot
```

```python
lag_plot(miles['MilesMM'], lag = 2)
```

# Auto-correlation Plots

```python
from pandas.plotting import autocorrelation_plot
```

```python
autocorrelation_plot(miles['MilesMM'])
```

* The Lags show the number of lags to the Original Values i.e. Lag(20) = y(t-20)

* Auto-correlation can be used to understand seasonality in the data. When seasonality is encountered, the auto-correlation plot shows a dip in the correlation values.
